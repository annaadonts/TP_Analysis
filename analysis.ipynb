```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12, 8))
img = plt.imread(image_path)
ax.imshow(img)
ax.axis('off')

ax.axvline(x=15, color='red', linestyle='--', linewidth=2, label='Proposed TP Lower Bound (15 pips)')
ax.axvline(x=20, color='red', linestyle='--', linewidth=2, label='Proposed TP Upper Bound (20 pips)')

ax.annotate("Suggested TP Range", xy=(17.5, 0.35), xytext=(30, 0.45),
            color="red", fontsize=12, arrowprops=dict(facecolor='red', shrink=0.05))

plt.legend(loc='upper right', fontsize=10)
plt.title("Adjusted Take Profit (TP) Range on EURUSD Plot", fontsize=14)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

swing_mean = 5.74
swing_median = 4.70
swing_std = 4.41

x = np.linspace(0, 20, 500)
swing_distribution = (1 / (swing_std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - swing_mean) / swing_std) ** 2)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, swing_distribution, label="Swing Distance Distribution", color="blue", linewidth=2)

ax.axvspan(15, 20, color="red", alpha=0.3, label="Proposed TP Range (15-20 pips)")

ax.axvline(swing_mean, color="green", linestyle="--", label="Swing Mean (5.74)")
ax.axvline(swing_mean + swing_std, color="purple", linestyle="--", label="1 Std Dev (10.15 pips)")
ax.axvline(swing_mean + 2 * swing_std, color="orange", linestyle="--", label="2 Std Dev (14.56 pips)")

ax.set_xlabel("Distance (pips)", fontsize=12)
ax.set_ylabel("Density", fontsize=12)
ax.set_title("Swing Distance Distribution with Proposed TP Range", fontsize=14)
ax.legend(fontsize=10)
plt.grid(alpha=0.3)
plt.tight_layout()

plt.show()
```
